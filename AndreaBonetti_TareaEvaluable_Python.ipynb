**Ejercicio 1. (2 puntos)** La criba de Eratstenes es un mtodo ideado por Eratstenes, para conseguir los nmeros primos entre el 1 y un determinado nmero entero n. Este ejercicio permite poner en prctica muchas de las sentencias de control que hemos visto.


La criba de Eratostenes se puede programar como un algoritmo que consiste en formar  una tabla con todos los nmeros naturales comprendidos entre 2 y n,  se van tachando los nmeros que no son primos de la siguiente manera: 

* Comenzando por el 2 (considerado primo), se tachan todos sus mltiplos.
* Comenzando de nuevo, cuando se encuentra un nmero entero que no ha sido tachado, ese nmero es declarado primo, y se procede a tachar todos sus mltiplos.
* El proceso se repite iterativamente. El proceso termina cuando el cuadrado del siguiente nmero confirmado como primo es mayor que n.

Con todo esto, 

* Crea una funcin denominada `Eratostenes(n)`, que devuelva los nmero primos inferiores o iguales a n. 

* El tipo computacional de n es entero, modifica la funcin anterior `Eratostenes_Err` y crea un nueva funcin  para que si se introduce un valor real como argumento de la funcin, no devuelva error y  muestre un mensaje por pantalla informando al usuario de que el argumento sea entero.

```python
#Solucin 1

def Eratostenes(n):
    
    numero = [i for i in range(2,n+1)] #Pongo todos los nmeros en una lista que empieza por el 2
   
    for index,div in enumerate(numero):
        #El proceso termina cuando el siguiente nmero confirmado como primo es mayor que la raz cuadrada de n
        if div>(n**(1/2)):   
            break    
        
        for value in numero[index+1:]: #Utilizo el ndice index de cada elemento div, para que la bsqueda de mltiples empiece con el siguiente elemento index+1
            if value%div ==0:         #busco los mltiples de div y los elimino directamente de la lista
                numero.remove(value)               
        
        #print(div)
    return numero
    
    
Eratostenes(100)
```
